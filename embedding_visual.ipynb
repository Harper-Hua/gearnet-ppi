```python
import numpy as np
import umap
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yeast_embeddings = pd.read_csv('/Users/harper.h/Documents/cs224w/final_project/output_rep.csv', header=None)
yeast_embeddings.head()
```

```python
reducer = umap.UMAP()
embedding = reducer.fit_transform(yeast_embeddings)
plt.figure(figsize=(10, 7))
plt.scatter(embedding[:, 0], embedding[:, 1])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Protein Embeddings', fontsize=12)
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Apply PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(yeast_embeddings)

# Plotting
plt.figure(figsize=(10, 7))
plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.title('PCA of Protein Embeddings')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Apply t-SNE
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(yeast_embeddings)

# Plotting
plt.figure(figsize=(10, 7))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1])
plt.title('t-SNE of Protein Embeddings')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.show()
```
